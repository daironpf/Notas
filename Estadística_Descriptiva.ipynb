# Una **Población** es un conjunto de personas u objetos en general, de los cuales nos interesa estudiar algunas características.

**Ejemplos:**
* Conjunto de votantes(podria interesarnos su preferencia electoral).
* Conjunto de automóviles en una ciudad (nos podria interesar el estado fisico de un automovil o sus niveles de emision de contaminantes).

* Conjunto de tornillos producidos por una cierta maquinaria en un día (estudiar si cumplen con cierto nivel de calidad).

# Una **Muestra** es caulquier subconjunto de una población.
> Al número de elementos de la muestra se le llama **tamaño de la muestra** y usaremos la letra **n** para denotarlo.

**Ejemplos:**
* Un 5% de un conjunto de votantes tomados al azar.
* El subconjunto de automoviles de color rojo.
* El subconjunto de animales de cierta especie en edad de reproducción.
* El subconjunto de tornillos defectuosos.

> Las muestras no son arbitrarias, se toman al azar con el fin de ser "representativas" de la población.

>Un **censo** es un registro de las características de interés de **todos** los elementos de una población

>Una **unidad de observación** es un grupo de elementos de una población del cual es posible obtener su información.

# Variable
Una **Variable** es una cracterística de nuestro interés que posee cada elemento de una población y que podemos medir. Cuya respuesta varía de un elemento a otro de nuestra problación.

**Ejemplos:**
* Población = "Conjunto de personas", Variable = "Edad" Cuál es su edad en años?, Variable ="Fumador" Usted Fuma?
* Población = "Conjunto de tornillos", Variable = "Longitud" Cuál es su longitud?, Variable = "TieneCalidad" Ha pasado usted el control de calidad?

De acurdo al tipo de respuestas que se obtenga las variables pueden ser:
* **Cuantitativas:** lo és si sus valores son números y representan una cantidad.
* **Cualitativas:** lo és si sus valores representan una cualidad, atributo o categoría. (también se les reconoce como **variables categóricas**)

> **Cuantitativas:** Se pueden clasificar en **Discretas** o **Continuas**; 
* Es **Discreta** si el conjunto de sus valores tiene un número finito de elementos o bien es infinito pero "numerable". 
* Es **Continua** si puedes tomar todos los valores de un intervalo **(a,b)** de números reales.

> **Cualitativa:** Se pueden clasificar en **Escala nominal** o **Escala ordinal**;
* Es **Nominal** si **no existe** un orden entre sus valores. (se dice que tal variable se mide mediante una escala nominal), Ej: Sexo, Religion, Estado Civil
* Es **Ordinal** si **existe** un orden entre sus valores.(se dice que tal variable se mide mediante una escala ordinal), Ej: Nivel ={Nulo, Básico, Medio, Avanzado} , EstadoDelArticulo = {Malo, Regular, Bueno}
* Una variable **Ordinal** puede ser considerada **Nominal** cuando se ignora o no se toma como relevante el orden entre los valores de la variable.

# Escalas de medición: variables cuantitativas

La escala utilizada es de **Intervalo** o de **Razón**:

* **Intervalo(se mide mediante una escala de intervalo):** si existe una noción de distancia entre dos valores pero no necesariamente se pueden efectuar operaciones aritméticas, el posible valor cero es arbitrario y no indica necesariamente la ausencia total de medida.
  * Ejemplos: Tiempo, Temperatura en C
* **Razón(se mide mediante una escala de razón):** si sus valores tienen un sentido físico, existe el cero absoluto que indica la ausencia total de medida, y se puede realizar cualquier operación aritmética(+,-,*,/).
  * Ejemplos: Altura, Peso, Distancia, Edad, Salario

> Toda variable con escala de medición de razón puede considerarce de intervalo cuando no es necesario llevar a cabo operaciones aritméticas entre sus valores.

# Desagrupamiento de Datos
Hay situaciones donde los datos disponibles para una o varias variables están agrupados en clases y no se tienen las observaciones individuales si se decea contar con los datos individuales estos pueden obternerse únicamente de manera aproximada:

**Ejemplo:** veremos unicamente variables cuantitativas; 
Una **variable X** es una pregunta o medición que se realiza a todos los elementos de una muestra (de tamaño n)

Supondremos que las **n** respuestas x1,x2,...,xn no se conocen, sólo se conoce cuántos de estos valores pertenecen a cada una de las clases de agrupaciones: C1,C2,...,Ck

Es decir; los datos se encuentran agrupados en las k clases.
El Problema es el siguiente: Cómo recuperar los valores x1,x2,...,xn a partir de las frecuencias de las clases ??

>Ejemplo: Supongamos que nos interesa estudiar la variable **"edad"** en años cumplidos en una población de seres humanos: 

definamos las clases de valores: 
C1 = {0,1,...,15} = "Niñez"
C2 = {16,17,...,39} = "Juventud"
C3 = {40,41,...} = "Vejez"
Datos: 

C1 = 4
C2 = 12
C3 = 6

> Como datos del problema, suponemos que la cardinalidad o numeros de elementos en la clase C1 es de 4, el de observaciones en la C2 es de 12 y tenemos 6 observaciones en el C3, para un total de 22 observaciones (n=22).

Nuestro problema es entonces el siguiente: ¿Cómo podemos aproximar las edades x1,x2,...,x22?
**Solución 1:** Se elige un dato representante de cada clase y se toma ese datos tantas veces como indica la frecuencia de la clase. Al dato representante escogido se le llama **marca de clase.**

**Solución1:**
> Por ejemplo para una variable continua si una clase de valores c = [a,b] es decir c está definida como el intervalo de a hacia b, entonces una marca de clase sería el punto medio del intervalo (a+b)/2 así se considera que sea observado este valor tantas veces como indica la frecuencia de la clase.

Regresando al ejemplo de las edades y escogiendo los valores de forma supuesta tendríamos que:
7: representante de C1
25: representante de C2
55: representante de C3

Nos daría estos Datos aproximados:
7,...,7 (4 veces) 
25,...,25 (12 veces)
55,...,55 (6 veces)

**Solución2:**
> Si una clase C tiene **m** observaciones, se escogen *m* valores "equiespaciados" (en la medida de lo posible) dentro de C.
Por ejemplo para una variable continua y si la clase C está definida como el intervalo [a,b] y si el numero de observaciones m = 3 entonces se escoge como datos los tres valores x espaciados que se encuentren dentro del intervalo (a,b) como se muestra en el diagrama.

3,6,9,12: Datos en C1
17,18,20,...: Datos en C2

> De esta forma hemos recuperado 22 observaciones individuales aproximadas

# Estadística Descriptiva:
La **Estadística Descriptiva** es un conjunto de técnicas y procedimientos que ayudan a describir, mostrar y resumir, la información de un conjunto de datos.

Los Estudios que se efectuán en la estadística descriptiva buscan describir las caractéristicas de la muestra utilizando **Números**, **Tablas** y **Gráficas** para realizar dichas descripciones.

#Estadística Inferencial
En la **Estadística Inferencial**: Se busca extender la información obtenida de una muestra a la población completa, en este caso es importante contar con una muestra representativa de la población y no existe seguridad completa de la valides de las afirmaciones, estas pueden enunciarse únicamente con un limitado grado de confiabilidad.

#Descripciones numéricas
Cantidad que se pueden calcular para describir algunas características de un conjunto de datos principalmente datos númericos.
Aquellas cantidades que buscan un valor central que represente a los datos se les llama **Medidas de Tendencia Central o Medidas de localización:**

*   Media
*   Moda
*   Mediana

Otras cantidades que han sido definidas con el objetivo de cuantificar la dispersion de un conjunto de datos se les llama **Medidades de Dispersión:**

*   Varianza
*   Desviación(Desv) Estándar
*   Desv. Media
*   Rango

#Media
**Definición:** Sean $x_{1},...,x_{n}$ observaciones o mediciones de una variable cuantitativa. La **media** es el número $\bar{x}=\frac{x_{1}+x_{2}+...+x_{n}}{n}$ se define como el promedio de los **n** números, la **media o media aritmética** es un valor promedio que resume y representa al conjunto de datos, está es la Medida de Localización más utilizada y muchas decisiones importantes son tomadas con base en este cantidad.

**Ejemplo:** Supongamos que queremos estudiar la variable **"Estatura"** en una población de seres humanos y contamos con una muestra de n=4 personas.
Y tenemos como observaciones de las variables estos cuatro números:
x1 = 1.70, x2 = 1.85, x3 = 1.65, x4 = 1.70
Entonces al allar el promedio con la ecuacion anterior de estos cuatros números nos da como resultado: 1.725 que sería la Estatura promedio de esta muestra de cuatro personas.

**Transformación 1:**
Se a cada uno de los datos x1,...,xn se le suma una constante **c**, entonces la media de los datos $x_{1}+c,...,x_{n}+c$ es $\bar{x}+c$ es decir sería la media de los datos originales más la constante **c**. 
>Está transformación corresponde a una traslación de los datos en **c** unidades.

**Transformación 2:**
Si a cada uno de los datos x1,...,xn se le multiplica por una constante **a** entonces la media de los datos $ax_{1},...,ax_{n}$ es $a\bar{x}$. 
>Está transformación equivale a efectuar un cambio de escala en la medición de los datos originales en **a** unidades.

# Moda
Se puede calcular tanto para valores númericos como no númericos.

**Definición:** Sean $x_{1},...,x_{n}$ observaciones o mediciones de una variable cuantitativa o cualitativa. La **moda** es el valor que aparece con mayor frecuencia en este conjunto de datos, si lo hubiera.

**Ejemplo Cuantitativo:** Nos importa la variable "Edad" en años cumplidos en una población de seres humanos y tenemos estas observaciones: 
$x_{1}=15, x_{2}=20, x_{3}=15, x_{4}=20, x_{5}=25, x_{6}=15$

La moda es el valor: **15**, de esta forma la moda es una medida de tendencia central que indica el valor observado con mayor frecuencia.

**Ejemplo Cualitativo:** Nos interesa la variable "Calidad" en una población de ciertos productos:
$x_{1}=Malo, x_{2}=Bueno, x_{3}=Bueno, x_{4}=Regular, x_{5}=Malo, x_{6}=Bueno$

La moda es el valor: **Bueno**.

**Propiedades Generales:**
1. La moda puede no existir.
2. La moda puede existir y ser única.
3. Pueden existir dos o más modas.(en este caso se dice que el conjunto de datos es bi-modal o multi-modal)
4. La moda permanece sin cambio cuando se añaden u omiten detos de "baja frecuencia"

**Transformación 1:** Si a cada uno de los datos numéricos $x_{1},...,x_{n}$ se le **suma** una constante **c**, entonces la moda de los datos $x_{1}+c,...,x_{n}+c$ es **$moda(x)+c$**. Esto presupone la existencia y unicidad de la moda para el conjunto de datos originales como moda(x).

>**comprobación:** La configuración de frecuencia de los datos transformados es la misma configuración de frecuencias de los datos originales. Lo único que cambia son los datos.

**Transformación 2:** Si a cada uno de los datos numéricos $x_{1},...,x_{n}$ se le **multiplica** una constante **a**, entonces la moda de los datos $ax_{1},...,ax_{n}$ es **$a*moda(x)$**. Esto presupone la existencia y unicidad de la moda para el conjunto de datos originales como moda(x).

>**comprobación:** Si $a\neq 0$, la configuración de frecuencias de los datos transformados es la misma configuración de frecuencia de los datos originales. Lo único que cambia son los datos. Si a=0, la moda es el valor 0=a*moda(x).




# Mediana

**Definición:** Sean $x_{1},...,x_{n}$ observaciones o mediciones de una variable cuantitativa(datos numéricos). Sean $x_{(1)}\leqslant x_{(2)} \leqslant ... \leqslant x_{(n)} $ los datos ordenados de menor a mayor, incluyendo repeticiones, los parentices en los índices indican que los datos están ordenados.
> Si **n** es par, entonces la **mediana** es $\tilde{x} = \frac{x(n/2)+x(n/2+1)}{2} $ ; Esta cantidad es el promedio aritmético entre el dato ordenado en la posición (n/2) y el dato ordenado en la posición (n/2+1) en este caso estos dos valores son los valores centrales del conjunto de datos.

**Ejemplo:** Supongamos que n=4,entonces tenemos lo siguiente $X_{(1)}\leqslant X_{(2)}\leqslant X_{(3)}\leqslant X_{(4)}$.

Como 4 es un número par por lo cual claramente tenemos dos números centrales aquí $X_{(2)}$ y $X_{(3)}$ el promedio de estos dos números es la mediana.

**Ejemplo:** Cuando n es un número impar, entonces la **mediana** es $\tilde{x} = x(\frac{n+1}{2})$, al ser n impar existe una observación que se encuentra justamente a la mitad de los datos ordenados, esta observación es el dato ordenado que se encuentra en la posición $\frac{n+1}{2}$.

Entonces tenemos la siguiente secuencia: $X_{(1)}\leqslant X_{(2)}\leqslant X_{(3)}\leqslant X_{(4)}\leqslant X_{(5)}$. y al aplicarle la ecuación mencionada tendría que $X_{(3)}$ sería el punto central de los datos ordenados. 

>De esta forma según sea el caso la mediana será el dato central o el promedio de los dos datos centrales. No existe una notación estandard para esta cantidad por lo cual usaremos la expresión (x tilde) $\tilde{x}$ para denotarla.

**Ejemplos del Cálculo de la Mediana:**
Consideremos que nos interesa estudiar la variable: estatura medida en cm. y supongamos que contamos con estas cinco observaciones:

a) n=5

165,165,**170**,172,174

$\tilde{x}= 170$

b) n=6

165,165,**170,172**,174,182

$\tilde{x}= \frac{170+172}{2}=171$

**Propiedades:**
1. La mediana puede ser uno de los datos observados o no serlo.
2. La mediana es un valor central que separa al conjunto de datos en dos partes, ambas con el mismo número de observaciones.
3. La mediana no cambia si algunas observaciones cambian de valor, siempre y cuando estos cambios ocurran dentro de una misma mitad de valores.

>Ahora veremos que ocurre con la Mediana bajo tranformaciones lineales de los datos.

**Transformación1:**
Si a cada uno de los datos numéricos $X_{1},...,X_{n}$ se le suma una constante c, entonces la mediana de los datos $X_{1}+c,...,X_{n}+c$ es $\tilde{X}+c$.

**Comprobación:**
Los datos transformados y ordenados de menor a mayor son $X_{(1)}+C\leqslant X_{(2)}+C\leqslant ... \leqslant X_{(n)}+C$ y esto es justamente la mediana más la constante C. 

Si n es impar, el dato central es: $X(\frac{n+1}{2})+C = \tilde{X}+C$

si n es par, el promedio de los dos datos centrales es = $\tilde{X}+C$

**Transformación2:**
Si a cada uno de los datos numéricos $X_{1},...,X_{n}$ se le multiplica una misma constante a, entonces la mediana de los datos transformados $aX_{1},...,aX_{n}$ es $a\tilde{X}$.

**Comprobación:**
Supongamos que la constante $a\geqslant0$ el argumento que presentaremos funciona de manera similar para el caso de -a. Los datos transformados y ordenados de menor a mayor son:

 $aX_{(1)}\leqslant aX_{(2)}\leqslant ... \leqslant aX_{(n)}$

**Si n es impar** el dato central es:  $ax(\frac{n+1}{2}) = a\tilde{x}$

**Si n es par** el dato central es: $a\tilde{x}$
